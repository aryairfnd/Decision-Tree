```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 1. Buat dataset
data = {
    'Jam_Belajar': [5, 3, 8, 2, 7, 4, 6, 1],
    'Jumlah_Absen': [1, 2, 0, 5, 1, 3, 2, 6],
    'Lulus': ['Ya', 'Tidak', 'Ya', 'Tidak', 'Ya', 'Tidak', 'Ya', 'Tidak']
}

df = pd.DataFrame(data)

# 2. Pisahkan fitur dan target
X = df[['Jam_Belajar', 'Jumlah_Absen']]
y = df['Lulus']

# 3. Split data train dan test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 4. Buat model Decision Tree
model = DecisionTreeClassifier(random_state=42)

# 5. Latih model
model.fit(X_train, y_train)

# 6. Prediksi data test
y_pred = model.predict(X_test)

# 7. Evaluasi akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi model Decision Tree: {accuracy:.2f}")

# 8. (Opsional) Tampilkan prediksi dan data sebenarnya
result = pd.DataFrame({'Aktual': y_test, 'Prediksi': y_pred})
print(result)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# 1. Dataset
data = {
    'Jam_Belajar': [5, 3, 8, 2, 7, 4, 6, 1],
    'Jumlah_Absen': [1, 2, 0, 5, 1, 3, 2, 6],
    'Lulus': ['Ya', 'Tidak', 'Ya', 'Tidak', 'Ya', 'Tidak', 'Ya', 'Tidak']
}
df = pd.DataFrame(data)

# 2. Fitur dan target
X = df[['Jam_Belajar', 'Jumlah_Absen']]
y = df['Lulus']

# 3. Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 4. Model Decision Tree
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# 5. Prediksi dan evaluasi
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi: {accuracy:.2f}")

# 6. Visualisasi pohon keputusan
plt.figure(figsize=(12,8))
plot_tree(model,
          feature_names=['Jam_Belajar', 'Jumlah_Absen'],
          class_names=model.classes_,
          filled=True,
          rounded=True,
          fontsize=14)
plt.show()
```
